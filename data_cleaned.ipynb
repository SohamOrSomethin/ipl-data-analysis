```python
import pandas as pd
import numpy as np


df = pd.read_csv(r"C:\Users\deole\Desktop\python\ipl\IPL.csv",low_memory=False)
# print(df.columns)
# print(df.head())
```

```python
numeric_cols = ['runs_batter', 'balls_faced', 'runs_total', 'runs_bowler',
                'team_runs', 'team_balls', 'team_wicket', 'batter_runs',
                'batter_balls', 'bowler_wicket']
```

```python
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

df.replace(r"^\s*$", np.nan, regex=True, inplace=True)
```

```python
print(df.isna().sum())  # show missing‑value count per column
print(df.info())

print(df.info())
print(df.shape)
```

```python
df["batting_team"] = df["batting_team"].str.strip().str.title()
df["bowling_team"] = df["bowling_team"].str.strip().str.title()
df["batter"] = df["batter"].str.strip()
df["bowler"] = df["bowler"].str.strip()
```

```python
df["is_four"] = (df["runs_batter"] == 4)
df["is_six"] = (df["runs_batter"] == 6)
```

```python
df["legal_ball"] = (df["extra_type"].isna()) & (df["valid_ball"] == 1)
```

```python
df = df.set_index(["match_id", "innings"])
```

```python
#most runs by  batsmen
```

```python
batter_runs = (
    df.groupby("batter", as_index=False)
      ["runs_batter"]
      .agg(["sum", "count"])
      .rename(columns={"sum": "total_runs", "count": "balls_faced"})
)

batter_runs["strike_rate"] = (batter_runs["total_runs"] / batter_runs["balls_faced"]) * 100

top_batters = (
    batter_runs
    .sort_values("total_runs", ascending=False)
    .head(10)
)
print("Top 10 batters by total runs")
print(top_batters)
```

```python
#most sixes by career
```

```python
sixes = (
    df[df["is_six"]]
    .groupby("batter", as_index=False)
    ["is_six"]
    .sum()
    .rename(columns={"is_six": "sixes"})
    .sort_values("sixes", ascending=False)
    .head(10)
)

print("Top 10 batters by sixes")
print(sixes)
```

```python
# Most fours by batter
```

```python
fours = (
    df[df["is_four"]]
    .groupby("batter", as_index=False)
    ["is_four"]
    .sum()
    .rename(columns={"is_four": "fours"})
    .sort_values("fours", ascending=False)
    .head(10)
)

print("Top 10 batters by fours")
print(fours)
```

```python
# Most wickets by bowler
```

```python
wickets = (
    df[df["wicket_kind"].notna()]
    .groupby("bowler", as_index=False)
    ["bowler_wicket"]
    .sum()
    .rename(columns={"bowler_wicket": "wickets"})
    .sort_values("wickets", ascending=False)
    .head(10)
)

print("Top 10 bowlers by wickets")
print(wickets)
```

```python
#most runs by team
```

```python
team_runs = (
    df.groupby("batting_team", as_index=False)
      ["runs_total"]
      .sum()
      .sort_values("runs_total", ascending=False)
      .head(10)
)

print("Top 10 teams by total runs scored")
print(team_runs)
```

```python
#Most sixes by team
```

```python
team_sixes = (
    df[df["is_six"]]
    .groupby("batting_team", as_index=False)
    ["is_six"]
    .sum()
    .rename(columns={"is_six": "sixes"})
    .sort_values("sixes", ascending=False)
    .head(10)
)

print("Top 10 teams by sixes")
print(team_sixes)
```

```python
#Season‑wise stats 
```

```python
season_top_batter = (
    df.groupby(["season", "batter"], as_index=False)
      ["runs_batter"]
      .sum()
)

season_top_batter = season_top_batter.sort_values(
    ["season", "runs_batter"], 
    ascending=[True, False]
)

top_per_season = (
    season_top_batter
    .groupby("season")
    .head(1)
    .reset_index(drop=True)
)

print("Top run‑scorer in each season")
print(top_per_season)
```
